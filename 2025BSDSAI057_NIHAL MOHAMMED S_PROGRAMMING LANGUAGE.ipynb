```python
# PROBLEM 1
# PART 1
def gcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m

print(gcd(45, 25))
print(gcd(100, 450))
print(gcd(7, 3))
print(gcd(1560, 2980))
```

```python
# PART 2
print(gcd(-56, 156))
print(gcd(56, -156))
print(gcd(-56, -156))
"""
In the case of either of the first number or second number being the only negative input, the output depends on the least value
given as an input. For example, If -5 and 45 are the inputs, the output will be positive. If 5 and -45 are the inputs, the output
will be negative.
"""
```

```python
# PROBLEM 2
# PART 1
def euler_limit(x):
    y = 1 + (1 / x)
    z = y ** x
    return float(z)
num = int(input("Please enter the number:"))
euler_limit(num)
```

```python
# PART 2
'''
When we put in values which are very huge, the output is given as "1.0"
This can be explained using the Equation(2).
When 'n' is too large, the value of (1/x) becomes very negligible to the point where it can be considered as zero.
Which would leave us only with 1 to the power of x which is equal to 1.
Hence, the output will be "1.0".
'''
```

```python
# PART 3
def factorial(x):
    if x == 0:
        return 1
    else:
        return x * factorial(x-1)

def euler_sum(y):
    total_summation = 0
    for i in range(y+1):
        b = factorial(i)
        total_summation += 1/b
    return float(total_summation)

num1 = int(input("Please enter the first number:"))
num2 = int(input("Please enter the second number:"))

factorial(num1)
euler_sum(num2)
```

```python
# PART 4
import math

def euler_approx(z: float):
    final_answer = 0
    i = 0
    while math.exp(1) - euler_sum(i) > z:
        i = i + 1
        final_answer = euler_sum(i)
    return final_answer

num = float(input("Please enter the number:"))
print(euler_approx(num))
```

```python
# PART 5
import math

n = int(input("Enter a number: "))

def print_euler_sum_error(n):
    for i in range(1, n+1):
        print(f"for i = {i}, {euler_sum(i) - math.exp(1)}")

def print_euler_lim_error(n):
    for i in range(1, n+1):
        print(f"for i = {i}, {euler_limit(i) - math.exp(1)}")

print_euler_sum_error(n)
print()
print_euler_lim_error(n)
```

```python
# PROBLEM 3
# PART 1
def is_power_of_2(n):
    if n == 1:
        return True
    if n % 2 != 0:
        return False
    return is_power_of_2(n // 2)

print(is_power_of_2(1))
print(is_power_of_2(2))
print(is_power_of_2(4))
print(is_power_of_2(16))
print(is_power_of_2(64))
print(is_power_of_2(3))
print(is_power_of_2(6))
print(is_power_of_2(100))
```

```python
# PART 2
def is_power(b, n):
    if n == 1:
        return True
    if n % b != 0:
        return False
    return is_power(b, n // b)

print(is_power(2, 8))
print(is_power(2, 16))
print(is_power(3, 27))
print(is_power(5, 125))
print(is_power(3, 10))
print(is_power(2, 10))
print(is_power(10, 1000))
print(is_power(2, 64) == is_power_of_2(64))
```
